DengAI: Initial Thoughts
===========

Baseline
------

I will train a linear regression model on the features that have the highest pearson correlation with the total dengue cases. Since the data given is for two areas that have different climates, and temperature and humidity are key elements for the size of the  mosquito population, I will create two separate linear regression models. For each model, I will calculate the pearson correlation and pick 2-3 features with the highest values. From there, I will use these features to create a linear regression line, serving as the baseline.  

Idea 1
------
While there is a correlation between humidity and mosquitoes, I hypothesize that the reason is because mosquitoes need water to lay eggs, and thus when it is humid, it may also imply that there are bodies of water readily available for breeding. However, using the humidity levels of the day to predict the total cases of dengue may not be acurrate as mosquitoes take 10-14 days to fully develop and be able to feed and thus transmit the fever. It also takes ~10 days for symptoms to develop. Thus I shifted <i>reanalysis_specific_humidity_g_per_kg</i> up by  3 blocks (21 days) as a new feature; this may result in a higher correlation. (This is the new feature I calculated)

Idea 2
------
Similarly I will also attempt to shift all of the climate variables as well by 1 block, 2 block, and 3 blocks and see which of the shifts will result in a higher correlation. Higher Temperatures result in more active feeding days for mosquitoes, which may expedite the dengue fever transmittion process.

Idea 3
------
It may be possible to scrape mosquito population data from the web for San Juan and Iquitos and use that data as a feature since mosquito population should have a high correlation with the total dengue fever cases. However, this data would also need to be shifted up 1 block (1-2 weeks) as symptoms occur in 10 days
